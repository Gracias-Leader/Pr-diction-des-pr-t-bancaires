- ### Type de problèmes:
  * Regroupement et ciblage des clients en fonction des prêts bancaires accorder lier a leur informatios personnelles.
 
- ### Type de modèle utilisé:
  * Modèle de Machine learning non supervisé : `K-Means Cluster`

## 3 Traitement des valeurs manquantes et des valeurs aberrantes

```python
# supprimer la colonne "Loan_ID"
df.drop("Loan_ID", axis=1, inplace=True)
```

```python
# Les valeurs manquantes
df_missing = (df.isnull().sum()/df.shape[0]).sort_values(ascending=False)
df_missing
```

```python
# Voir une carte des valeurs manquantes
plt.figure(figsize=(15,5))
sns.heatmap(df.isnull(), cbar=True, cmap="viridis")
plt.title("La carte montrant les valeurs nulles de mon dataset", fontsize=20)
plt.show()
```

```python
# Les colonnes qui contiennes les valeurs catégorielles
df["Credit_History"] = df["Credit_History"].fillna(df["Credit_History"].value_counts().idxmax())
df["Self_Employed"] = df["Self_Employed"].fillna(df["Self_Employed"].value_counts().idxmax())
df["Gender"] = df["Gender"].fillna(df["Gender"].value_counts().idxmax())
df["Dependents"] = df["Dependents"].fillna(df["Dependents"].value_counts().idxmax())
df["Married"] = df["Married"].fillna(df["Married"].value_counts().idxmax())

# Les colonnes qui contiennes les valeurs numériques
df["Loan_Amount_Term"] = df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mean())
df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].mean())
```

```python
# Virifier si j'ai encore des valeurs nulle dans mon data frame
df.isnull().sum()
```
