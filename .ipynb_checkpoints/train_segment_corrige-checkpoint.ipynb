## 1 BUSINESS UNDESTANDING (Énoncé du problème)
- Problème d'identification des segments de clients éligibles aux différents montants de prêt, afin de cibler   précisément ces clients.

- ### Objectif:
  * L'entreprise souhaite automatiser le processus d'éligibilité aux prêts (en temps réel) en se basant sur les informations fournies par le client lors du remplissage du formulaire de demande.
 
### Partie A : Segmentation (Non supervisée )

Objectif : Stratégie marketing et commerciale.

Modèle : `K-Means`.
  * Regroupement et ciblage des clients en fonction des prêts bancaires accorder lier a leur informatios personnelles.

   * Résultat : Vous créez des "personas" de clients pour offrir des produits adaptés (ex: proposer des micro-crédits au segment qui demande de petits montants).
 
 
### Partie B : Prédiction d'éligibilité (Supervisée)

`Objectif`:Automatisation des décisions de crédit. 
- Modèle Classification (Régression Logistique, Random Forest et XGBoost).Variable cible ($Y$) : `Loan_Status`. 
- Variables prédictives ($X$) : Le revenu, l'historique de crédit, le montant demandé, le statut marital, etc.

* ### ENCODER LES VARIABLES 

```python
# Variables binaires → map direct
binary_cols = {
    "Gender":      {"Male": 1, "Female": 0},
    "Married":     {"Yes": 1, "No": 0},
    "Education":   {"Graduate": 1, "Not Graduate": 0},
    "Self_Employed": {"Yes": 1, "No": 0},
    "Loan_Status": {"Y": 1, "N": 0}
}

for col, mapping in binary_cols.items():
    df[col] = df[col].map(mapping)

# Variables à 3+ modalités → get_dummies
multi_cols = ["Dependents", "Property_Area"]
df = pd.get_dummies(df, columns=multi_cols, drop_first=True)

# Convertir en float
df = df.astype("float")
```

## 5 Les test statistiques 

```python
# Tests statistiques — Mann-Whitney U (variables continues vs Loan_Status)
alpha = 0.05
Resultat = []
for col in df.columns.drop("Loan_Status"):
    g0 = df[df["Loan_Status"] == 0][col]
    g1 = df[df["Loan_Status"] == 1][col]
    stat, p = mannwhitneyu(g0, g1, alternative="two-sided")
    Resultat.append({
        "Variable": col,
        "Test": "mannwhitneyu",
        "statistique": round(stat, 4),
        "P_value": round(p, 4),
        "Resultat": "Significative" if p < 0.05 else "Non significative"})

df_mw = pd.DataFrame(Resultat)
df_mw
```

## 6 DATA PREPROCESSING

```python
# Preparation des données en variables indépendantes et  cible
X = df.drop("Loan_Status", axis=1)
y = df["Loan_Status"]

# Separation des données en train test
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=40)
```

```python
# Pipeline des modèles de ML supervisé)

# La pipeline  Pour la régression logistic
pipe_lr = Pipeline([("Scaler", StandardScaler()),("Logistic", LogisticRegression())])
```

```python
# La pipeline  Pour la régression logistic
pipe_rf = Pipeline([("Scaler", StandardScaler()),("Random", RandomForestClassifier(n_estimators=100, max_depth=5,class_weight="balanced", random_state=42))])
```

```python
# La ration 
ratio = (y_train == 0).sum() / (y_train == 1).sum()

# Pipeline du XGBoost
pipe_xgb = Pipeline([("Scaler", StandardScaler()),("xgboost", XGBClassifier(scale_pos_weight=ratio, eval_metric="logloss", random_state=42))])
```

```python
# Entrainement des modèles
pipe_lr.fit(X_train,y_train)
pipe_rf.fit(X_train,y_train)
pipe_xgb.fit(X_train,y_train)
```

```python
# Validation croisée — scoring="recall" cohérent avec l'objectif
for name, pipe in [("logistic",pipe_lr),("randomforest",pipe_rf),("xgboost",pipe_xgb)]:
    scores = cross_val_score(pipe, X_train, y_train, cv=5, scoring="recall")
    print(f"{name}: recall_cv = {scores.mean():.4f} (+/- {scores.std():.4f})")
```

```python
# Faire des prédiction des modeles
y_pred_lr = pipe_lr.predict(X_test)
y_pred_rf = pipe_rf.predict(X_test)
y_pred_xg = pipe_xgb.predict(X_test)
```

```python
# Voir la matrice de coufusion et la classification repport
List = [("logistic",y_pred_lr),
        ("randomforest",y_pred_rf),
        ("xgboost",y_pred_xg)]

for name, y_pred in List:
    print(f"\n==={name.upper()}===")
    print("La matrice de confusion:")
    print(confusion_matrix(y_test, y_pred))
    print("\nLa classification report:")
    print(classification_report(y_test, y_pred))
```

## 7 Hyperparameter tuning

```python

# Dictionnaire des pipelines
pipelines = {
    "Logistic": pipe_lr,
    "Random": pipe_rf,
    "xgboost": pipe_xgb}
```

```python
# Dictionnaire des hyperparamètres
param_grids = {

    "Logistic": {
        "Logistic__C": [0.1, 1, 10]
    },

    "Random": {
        "Random__n_estimators": [100, 200],
        "Random__max_depth": [3, 5, 10]
    },

    "xgboost": {
        "xgboost__n_estimators": [100, 200],
        "xgboost__max_depth": [3, 5],
        "xgboost__learning_rate": [0.01, 0.1]
    }}
```

```python
# GridSearchCV — scoring="recall" pour maximiser la détection des emprunteurs éligibles
best_estimators = {}  # stocke le meilleur pipeline pour chaque algo

for name, model in pipelines.items():
    grid = GridSearchCV(
        estimator=model,
        param_grid=param_grids[name],
        cv=5,
        scoring="recall",
        n_jobs=-1)
    grid.fit(X_train, y_train)
    best_estimators[name] = grid.best_estimator_

    print(f"\n=== {name.upper()} ===")
    print("Meilleurs parametres :", grid.best_params_)
    print("Meilleur recall (CV)  :", round(grid.best_score_, 4))
```

## 8 Selection finale du meilleur modele
Evaluation de chaque `best_estimator_` sur le test set avec Recall, F1 et AUC-ROC.

```python
# Tableau comparatif sur le test set
from sklearn.metrics import recall_score, f1_score, roc_auc_score

print(f"{"Modele":<22} {"Recall":<10} {"F1-score":<10} {"AUC-ROC"}")
print("-" * 55)

resultats_finaux = {}
for name, model in best_estimators.items():
    y_pred  = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    recall  = recall_score(y_test, y_pred)
    f1      = f1_score(y_test, y_pred)
    auc     = roc_auc_score(y_test, y_proba)
    resultats_finaux[name] = {"recall": recall, "f1": f1, "auc": auc}
    print(f"{name:<22} {recall:<10.4f} {f1:<10.4f} {auc:.4f}")

# Selection selon le recall (metrique principale pour le churn)
best_name  = max(resultats_finaux, key=lambda k: resultats_finaux[k]["recall"])
best_model = best_estimators[best_name]
print(f"\n-> Meilleur modele selectionne : {best_name}")
```
